```python
# DataFrame
df = pd.read_csv("dataset_draft2.csv")
df.drop([
         "median_diff_holds",
         "mean_diff_holds",
         "Unnamed: 0",
         "holds", 
         "Name", "Moves",
         "Holdsets",
         "Id",
         "lengths_of_moves",
         "Grade_thermo",
         "Grade_ohe",
         "numeric_grades"], axis=1, inplace=True)

# sort by grade

df = df.sort_values(by=['Grade_numeric'])
df = df.reset_index(drop=True)
df.drop(["Grade_numeric"], axis=1, inplace=True)
df.head(10)
```

```python
df.columns
```

```python
# OVER SAMPLING

data = df.values
X, y =  data[:, 1:], data[:, [0]]
# label encode the target variable
y = LabelEncoder().fit_transform(y)


# transform the dataset
strategy = {0:500, 1:500, 2:500, 3:500, 4:500, 5:500, 6:500,7:500,8:500,9:500}
oversample = SMOTE(sampling_strategy=strategy)
X, y = oversample.fit_resample(X, y)
# summarize distribution
counter = Counter(y)
for k,v in counter.items():
    per = v / len(y) * 100
    print('Class=%d, n=%d (%.3f%%)' % (k, v, per))
# plot the distribution
pyplot.bar(counter.keys(), counter.values())
pyplot.show()
```

```python
# split in to train test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)
```

```python
import pandas as pd
from sklearn.feature_selection import SelectFromModel

clf = RandomForestClassifier(n_estimators=500, random_state=0, n_jobs=-1)

feat_labels = ['number_of_holds', 'downwards_count', 'only_upwards_count',
       'only_sideways_count', 'yellow_count', 'black_count', 'white_count',
       'downwards_percent', 'black_percent', 'white_percent', 'yellow_percent',
       'b_w_percent', 'only_upwards_percent', 'mean_distance_of_holds',"last_move_diff"]

# Train the classifier
clf.fit(X_train, y_train)

# Print the name and gini importance of each feature
for feature in zip(feat_labels, clf.feature_importances_):
    print(feature)
```

```python
# Model

# labels
grade_to_num = {"6B+":0,"6C":1,"6C+":2,"7A":3,"7A+":4,"7B":5,"7B+":6,"7C":7,"7C+":8,"8A":9,"8A+":10,"8B":11}
num_to_grade = {v: k for k, v in grade_to_num.items()}

# model
weights = {0:2.0, 1:2.0, 2:2.0, 3:4.0, 4:4.0, 5:2.0, 6:2.0,  7:2.0, 8:2.0, 9:2.0, 10:1.0, 11:1.0}
model = RandomForestClassifier(max_depth=16,n_estimators=500, class_weight=weights)
model.fit(X_train, y_train)
y_pred = model.predict(X_train)

# classification report
print(classification_report(y_train, y_pred))


# confusion matrix
confusion_matrix = metrics.confusion_matrix(y_train, y_pred)
confusion_matrix_df = pd.DataFrame(confusion_matrix)
confusion_matrix_df.rename(num_to_grade,axis=0,inplace=True)
confusion_matrix_df.rename(num_to_grade,axis=1,inplace=True)


# plot confusion matrix
plt.figure(figsize=(10,8))
sns.heatmap(confusion_matrix_df, annot=True,fmt='g')
plt.title('Training Confusion Matrix')
plt.ylabel('Actal Values')
plt.xlabel('Predicted Values')
plt.show()
```

```python
# Test

# predict
y_pred = model.predict(X_test)

# classification report
print(classification_report(y_test, y_pred))

# confusion matrix
confusion_matrix = metrics.confusion_matrix(y_test, y_pred)
confusion_matrix_df = pd.DataFrame(confusion_matrix)
confusion_matrix_df.rename(num_to_grade,axis=0,inplace=True)
confusion_matrix_df.rename(num_to_grade,axis=1,inplace=True)


plt.figure(figsize=(10,8))
sns.heatmap(confusion_matrix_df, annot=True,fmt='g')
plt.title('Test Confusion Matrix',)
plt.ylabel('Actal Values')
plt.xlabel('Predicted Values')
plt.show()
```
