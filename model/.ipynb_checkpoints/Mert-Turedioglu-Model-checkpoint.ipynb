```python
import imblearn
from collections import Counter
from matplotlib import pyplot
import pandas as pd
from imblearn.over_sampling import SMOTE
from imblearn.under_sampling import RandomUnderSampler
import numpy as np
from numpy import std
from numpy import mean
from pandas import read_csv
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import RepeatedStratifiedKFold
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.datasets import make_classification
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import f_classif
from sklearn.feature_selection import SelectFromModel
from sklearn.metrics import classification_report
from sklearn import metrics
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# DataFrame
df = pd.read_csv("non_bench_and_bench.csv")
df.drop([
         "Unnamed: 0",
         "holds", 
        "Grade_numeric",
        "Moves",
         "Id",
         "lengths_of_moves",
         "Grade_thermo",
         "Grade_ohe",
         "numeric_grades",
         "IsBenchmark",
         "Repeats",
         "median_diff_holds",
        "mean_diff_holds",
        "Holdsets"
        ],
        axis=1, inplace=True)

# sort by grade


df = df.reset_index(drop=True)
df.drop(["Name"], axis=1, inplace=True)
df.drop(df.head(3).index,inplace=True)
df
```

```python
# split in to train test sets

data = df.values
X, y =  data[:, 1:], data[:, [0]]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=1)


y_train = LabelEncoder().fit_transform(y_train)
y_test = LabelEncoder().fit_transform(y_test)


# OVER SAMPLING


# transform the dataset
strategy = {0:212 , 1:101, 2:101, 3:115, 4:92, 5:75 , 6:75 ,7:59 ,8:50,9:18}
oversample = SMOTE(sampling_strategy=strategy)
X_train, y_train = oversample.fit_resample(X_train, y_train)
# summarize distribution

 # DOWN SAMPLING

#undersample = RandomUnderSampler(sampling_strategy="not minority")
#X_train, y_train = undersample.fit_resample(X_train, y_train)
#
counter = Counter(y_train)
for k,v in counter.items():
    per = v / len(y_train) * 100
    print('Class=%d, n=%d (%.3f%%)' % (k, v, per))
# plot the distribution
pyplot.bar(counter.keys(), counter.values())
pyplot.show()
```

```python
parameters = {
    "n_estimators":[5,10,50,100,250,500,1000],
    "max_depth":[2,4,8,16,32,None]
}

model = RandomForestClassifier()

from sklearn.model_selection import GridSearchCV
cv = GridSearchCV(model,parameters,cv=5)
cv.fit(X_train,y_train.ravel())

def display(results):
    print(f'Best parameters are: {results.best_params_}')
    print("\n")
    mean_score = results.cv_results_['mean_test_score']
    std_score = results.cv_results_['std_test_score']
    params = results.cv_results_['params']
    for mean,std,params in zip(mean_score,std_score,params):
        print(f'{round(mean,3)} + or -{round(std,3)} for the {params}')
display(cv)
```

```python
import pandas as pd
from sklearn.feature_selection import SelectFromModel

clf = RandomForestClassifier(n_estimators=1000, random_state=0, n_jobs=-1,max_depth=None)

feat_labels = ['number_of_holds', 'downwards_count', 'black_count',
       'white_count', 'downwards_percent', 'black_percent', 'white_percent',
       'mean_distance_of_holds', 'last_move_diff', 'first_move_diff']

# Train the classifier
clf.fit(X_train, y_train)

# Print the name and gini importance of each feature
for feature in zip(feat_labels, clf.feature_importances_):
    print(feature)
```

```python
# Model

# labels
grade_to_num = {"6B+":0,"6C":1,"6C+":2,"7A":3,"7A+":4,"7B":5,"7B+":6,"7C":7,"7C+":8,"8A":9,"8A+":10,"8B":11}
num_to_grade = {v: k for k, v in grade_to_num.items()}

# class weights
weights = {0:1.0, 1:4.0, 2:4.0, 3:4.0, 4:4.0, 5:4.0, 6:4.0,  7:5.0, 8:5.0, 9:5.0, 10:5.0, 11:1.0}

model = RandomForestClassifier(max_depth=10,n_estimators=8,max_features=None)
model.fit(X_train, y_train)
y_pred = model.predict(X_train)

# classification report
print(classification_report(y_train, y_pred))


# confusion matrix
confusion_matrix = metrics.confusion_matrix(y_train, y_pred)
confusion_matrix_df = pd.DataFrame(confusion_matrix)
confusion_matrix_df.rename(num_to_grade,axis=0,inplace=True)
confusion_matrix_df.rename(num_to_grade,axis=1,inplace=True)


# plot confusion matrix
plt.figure(figsize=(10,8))
sns.heatmap(confusion_matrix_df, annot=True,fmt='g')
plt.title('Training Confusion Matrix')
plt.ylabel('Actal Values')
plt.xlabel('Predicted Values')
plt.show()
```

```python
# Test

# predict
y_pred = model.predict(X_test)

# classification report
print(classification_report(y_test, y_pred))

# confusion matrix
confusion_matrix = metrics.confusion_matrix(y_test, y_pred)
confusion_matrix_df = pd.DataFrame(confusion_matrix)
confusion_matrix_df.rename(num_to_grade,axis=0,inplace=True)
confusion_matrix_df.rename(num_to_grade,axis=1,inplace=True)


plt.figure(figsize=(10,8))
sns.heatmap(confusion_matrix_df, annot=True,fmt='g')
plt.title('Test Confusion Matrix',)
plt.ylabel('Actal Values')
plt.xlabel('Predicted Values')
plt.show()
```

```python
def plus_minus_one_accuracy(confusion_matrix):
    total_sum = 0
    diagonal_sum = 0
    minus_one_sum = 0
    plus_one_sum = 0
        
    for row in confusion_matrix:
        total_sum += sum(row)
        
    for i in range(len(confusion_matrix)-1):
        minus_one_sum += confusion_matrix[i][i+1]
        
    for i in range(len(confusion_matrix)-1):
        plus_one_sum += confusion_matrix[i+1][i]
        
    for i in range(len(confusion_matrix)):
        diagonal_sum += confusion_matrix[i][i]
    print("+-1 Accuracy:")
    return (diagonal_sum + minus_one_sum + plus_one_sum)/total_sum
plus_minus_one_accuracy(confusion_matrix)
```
