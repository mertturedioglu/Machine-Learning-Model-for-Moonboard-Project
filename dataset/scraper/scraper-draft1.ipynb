```python
from selenium import webdriver
from selenium.webdriver.support.ui import Select
from bs4 import BeautifulSoup
import time
```

```python
# login

web = "https://www.moonboard.com/"
path = "C:/Users//mertt/Desktop/udemy/Scraping/chromedriver.exe"
driver = webdriver.Chrome(path)
driver.get(web)
driver.maximize_window()

login = driver.find_element_by_xpath('//*[@id="loginDropdown"]')
login.click()

time.sleep(1)
    
username = driver.find_element_by_xpath('//*[@id="Login_Username"]')
username.send_keys("???")

time.sleep(1)

password = driver.find_element_by_xpath('//*[@id="Login_Password"]')
password.send_keys("???")

time.sleep(1)

login_button = driver.find_element_by_xpath('//*[@id="navlogin"]')
login_button.click()

time.sleep(5)


# select moonboard 2016

moonboard2016 = Select(driver.find_element_by_id('Holdsetup'))
moonboard2016.select_by_value('1')

# pagination

time.sleep(5)

pagination = driver.find_elements_by_xpath('//ul[contains(@class, "k-pager-numbers k-reset")]')[1]
pages = pagination.find_elements_by_tag_name('li')
last_page = int(pages[-2].text)
current_page = 1

while current_page <= last_page:
    time.sleep(2)  
    containers = driver.find_elements_by_xpath("//*[@class='k-grid-content k-auto-scrollable']")
    routes = containers[1].find_elements_by_xpath('.//a')
    
    for route in routes:
        time.sleep(2)
        route.click()
        time.sleep(3)
        driver.switch_to.window(driver.window_handles[-1])
        time.sleep(3)
        pageSource = driver.page_source
        with open("page_source.txt", "a", encoding="utf-8") as f:
            f.write(pageSource)
            f.close
        driver.close()
        driver.switch_to.window(driver.window_handles[0])
    
    current_page = current_page + 1
    
    try:
        next_page = driver.find_element_by_xpath('//span[contains(@class, "nextButton ")]')
        next_page.click()
    except:
        pass
        
driver.quit()

```
