```python
from selenium import webdriver
from selenium.webdriver.support.ui import Select
from bs4 import BeautifulSoup
import time
```

```python
# login

web = "https://www.moonboard.com/"
path = "C:/Users//mertt/Desktop/udemy/Scraping/chromedriver.exe"
driver = webdriver.Chrome(path)
driver.get(web)
driver.maximize_window()

login = driver.find_element_by_xpath('//*[@id="loginDropdown"]')
login.click()

time.sleep(1)
    
username = driver.find_element_by_xpath('//*[@id="Login_Username"]')
username.send_keys("???")

time.sleep(1)

password = driver.find_element_by_xpath('//*[@id="Login_Password"]')
password.send_keys("???")

time.sleep(1)

login_button = driver.find_element_by_xpath('//*[@id="navlogin"]')
login_button.click()

time.sleep(5)


# select moonboard 2016

moonboard2016 = Select(driver.find_element_by_id('Holdsetup'))
moonboard2016.select_by_value('1')

time.sleep(15)
    
                
# page 6


page_6 = driver.find_elements_by_xpath("//*[@data-page='6']")[1]
page_6.click()
time.sleep(5)


                
# page 7
containers = driver.find_elements_by_xpath("//*[@class='k-grid-content k-auto-scrollable']")
page_7 = driver.find_element_by_xpath("//*[@data-page='7']")
page_7.click()
time.sleep(10)

routes = containers[1].find_elements_by_xpath('.//a')

for route in routes:
    route.click()
    driver.switch_to.window(driver.window_handles[-1])
    pageSource = driver.page_source
    with open("page7.txt", "a", encoding="utf-8") as f:
        f.write(pageSource)
        f.close
    driver.close()
    driver.switch_to.window(driver.window_handles[0])           
    
# page 8

page_8 = driver.find_element_by_xpath("//*[@data-page='8']")
page_8.click()
time.sleep(10)

routes = containers[1].find_elements_by_xpath('.//a')

for route in routes:
    route.click()
    driver.switch_to.window(driver.window_handles[-1])
    pageSource = driver.page_source
    with open("page8.txt", "a", encoding="utf-8") as f:
        f.write(pageSource)
        f.close
    driver.close()
    driver.switch_to.window(driver.window_handles[0])  
    
# page 9

page_9 = driver.find_element_by_xpath("//*[@data-page='9']")
page_9.click()
time.sleep(2)

routes = containers[1].find_elements_by_xpath('.//a')

for route in routes:
    route.click()
    driver.switch_to.window(driver.window_handles[-1])
    pageSource = driver.page_source
    with open("page9.txt", "a", encoding="utf-8") as f:
        f.write(pageSource)
        f.close
    driver.close()
    driver.switch_to.window(driver.window_handles[0])  

    
# page 10

page_10 = driver.find_element_by_xpath("//*[@data-page='10']")
page_10.click()
time.sleep(2)

routes = containers[1].find_elements_by_xpath('.//a')

for route in routes:
    route.click()
    driver.switch_to.window(driver.window_handles[-1])
    pageSource = driver.page_source
    with open("page10.txt", "a", encoding="utf-8") as f:
        f.write(pageSource)
        f.close
    driver.close()
    driver.switch_to.window(driver.window_handles[0])  


# page 11

page_11 = driver.find_element_by_xpath("//*[@data-page='11']")
page_11.click()
time.sleep(2)

routes = containers[1].find_elements_by_xpath('.//a')

for route in routes:
    route.click()
    driver.switch_to.window(driver.window_handles[-1])
    pageSource = driver.page_source
    with open("page11.txt", "a", encoding="utf-8") as f:
        f.write(pageSource)
        f.close
    driver.close()
    driver.switch_to.window(driver.window_handles[0])  
    
    
# page 12

page_12 = driver.find_element_by_xpath("//*[@data-page='12']")
page_12.click()
time.sleep(2)

routes = containers[1].find_elements_by_xpath('.//a')

for route in routes:
    route.click()
    driver.switch_to.window(driver.window_handles[-1])
    pageSource = driver.page_source
    with open("page12.txt", "a", encoding="utf-8") as f:
        f.write(pageSource)
        f.close
    driver.close()
    driver.switch_to.window(driver.window_handles[0])   
    
driver.quit()
```
