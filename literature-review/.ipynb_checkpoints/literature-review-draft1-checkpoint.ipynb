* #### Stapel, F. (2020) A Heuristic Approach to Indoor Rock Climbing Route Generation

##### * Aim:
The main aim of the paper is to generate novel boulder problems(climbing routes). The achieve this goal, they asked the questions: 
* To generate a new boulder problem, what heuristics one may follow?
* *  How can one classify climbing holds?
* * How can one classify climbing moves?
* * How particular heuristics alter the difficulty grade of a boulder problem?

* With the help of these heuristics, how can one set a new boulder problem?

* What standart one may follow to assess a quality of a boulder problem?

The training board they used in this research is Moonboard 2017 which is a 40$^{\circ}$ overhang artificial climbing wall.

It may seem that the aim of my project and this paper does not overlap. To achive their goal, they assessed the difficulty grades of climbing routes. The overlap is here. Although their main goal does not corrensponds to mine, this assessment makes it so. In addition to this, their heuristic approach is valuable here. This heuristic approach will help me to have good feature set.

#### * Methods
They used greedy tree algorithms to generate a tree which consists of novel boulder problems. In the literature, people used machine learning models to assess the difficulty of a problem. In this research, they think that a heuristic approach will be better, since they found previous work unsuccessful in the sense that their models couldn't assess the difficulty of holds as people do. 



To grow a tree, they classified climbing holds, climbing moves and studied influence of these heuristics on the grade of a problem. 
##### Classification of holds:
There are 198 different holds on this board. The classified these holds regarding these aspects:

* A difficulty rating of gripping particular hold. They gave a questionnary consists of verbal phrases to experienced climbers to rate how difficult to grip particular hold on a scale of 1 to 5. 
* The hold's rotational angle to the ground. They defined 8 angles: east, west, south, north, north-west, north-east, etc. 
* The primary type of the hold: edge, sloper, pinch, etc.
* The secondary type of the hold.

##### Classification of moves:
By watching the videos of people climbing moonboard problems, they summarized the moves made in climbing moonboard problems: using a side pull, 'normal' climbing move, dead-pointing, dyno-ing, using a heel hook, etc. 



#### Influence of heuristics on the difficulty of a problem
They listed the heuristics below to assess the difficulty of a problem:
* Longer move, harder problem
* Harder hold, harder problem
* Less footholds, harder problem
* More special climbing moves, harder problem

##### Score of a hold = 
Regarding classification of holds, experienced climbers agreed upon the difficulty rate of the hard and easy holds, yet their ratings did not overlap regarding holds neither easy nor hard. 

Score of a hold = (2 * difficulty score + 1 * hold score + 3 * move score + 3 * rotation score) / 9. Each score has own equation which can be seen in the paper.

##### * Results
They asked experienced climbers to evaulate the routes generated and benchmark moonboard routes: %80 of the benchmark routes have good flow, yet it was %45 for the routes generated by the algorithm. %75 of the benchmark routes are enjoyable, whereas %40 of the generated routes are enjoyable. 
The main reason behind for this is that the algorithm does not take into account of the full state of the climbers on the routes on each move. What I mean by the full state of the climbers is that, the algorithm does not keep track of the positions of the limbs of the climbers when climbing the route. This leads to awkward moves in generated routes. 
They compared their grades with people's opinion's on the routed generated. These were pretty close. They were successful to estimate the grades of the routes.

* #### Duh, Y. S., & Chang, R. (2021). Recurrent neural network for moonboard climbing route classification and generation

##### * Aim:
Their main aim is to develop a model that generate moonboard problems, DeepRouteSet. To achieve this, they develop a pipeline, BetaMove, that returns hand sequence of a given route. This sequence output fed into another model that returns the difficulty grade of a given route, GradeNet. 

##### * Methods
As I briefly mentioned in the aim part, the BetaMove pipeline is developed. It is used to both to train GradeNet, and DeepRouteSet. When given a route, BetaMove generates many possible hand sequences, and returns the one with best scores. Thety used beam search algorithm to find the one with the best score. With these hand sequence data and grade label, they trained both GradeNet, which is a model returns difficulty grade of a route, and DeepRouteSet, that generates novel routes with grade label.
They treated this problem of classifying and generating route as an NLP problem. Because of the sequential nature of the climbing routes, they preferred RNN algorithm. Like music generator models, DeepRouteSet learns the patterns between adjoint moves and using these patterns to create new routes. As like sentiment analysis, they used sequantial data of hand sequences like like a textual data and grade information like sentiments. The architectures of DeepRouteSet and GradeNet are below:
